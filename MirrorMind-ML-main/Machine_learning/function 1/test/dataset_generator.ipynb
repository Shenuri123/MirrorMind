```python
from numba import jit
from gensim.utils import simple_preprocess

@jit
def process_data(data, df):
    q1 = 'How would you describe your emotional state '
    q2 = 'How would you describe your overall outlook on the future '
    q3 = 'What is your overall perception of your accomplishments and failures in life '
    q4 = 'What level of self-worth do you believe you currently possess '
    q5 = 'How would you describe your emotions when considering the possibility of something bad happening to you '
    q6 = 'How do you feel about yourself when you make mistakes or fail '
    q7 = 'The extent of your current thoughts about self-harm or suicide '
    q8 = 'What best describes your current level of irritation or annoyance '
    q9 = 'how would you rate your ability and approach to making decisions '
    q10 = 'How do you feel about your current appearance and any changes you might have noticed over time '
    q11 = 'How would you describe your current level of motivation and ability to perform tasks '
    q12 = 'What is your current level of satisfaction with various aspects of your life '

    for val1, answer1 in data[q1].items():
        for val2, answer2 in data[q2].items():
            for val3, answer3 in data[q3].items():
                for val4, answer4 in data[q4].items():
                    for val5, answer5 in data[q5].items():
                        for val6, answer6 in data[q6].items():
                            for val7, answer7 in data[q7].items():
                                for val8, answer8 in data[q8].items():
                                    for val9, answer9 in data[q9].items():
                                        for val10, answer10 in data[q10].items():
                                            for val11, answer11 in data[q11].items():
                                                for val12, answer12 in data[q12].items():
                                                    score = int(val1) + int(val2) + int(val3) + int(val4) + int(
                                                        val5) + int(val6) + int(val7) + int(val8) + int(val9) + int(
                                                        val10) + int(val11) + int(val12)

                                                    new_row = [simple_preprocess(q1 + answer1),
                                                               simple_preprocess(q2 + answer2),
                                                               simple_preprocess(q3 + answer3),
                                                               simple_preprocess(q4 + answer4),
                                                               simple_preprocess(q5 + answer5),
                                                               simple_preprocess(q6 + answer6),
                                                               simple_preprocess(q7 + answer7),
                                                               simple_preprocess(q8 + answer8),
                                                               simple_preprocess(q9 + answer9),
                                                               simple_preprocess(q10 + answer10),
                                                               simple_preprocess(q11 + answer11),
                                                               simple_preprocess(q12 + answer12), score]
                                                    df.loc[len(df)] = new_row
                                                    print(df.shape)
                                                    if df.shape[0] > 500000:
                                                        df.to_parquet('data123.parquet')
                                                        

# Call the function
process_data(data, df)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Download NLTK resources if not already downloaded
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')

lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    words = nltk.word_tokenize(text)  # Tokenize the text into words
    words = [lemmatizer.lemmatize(word) for word in words]  # Lemmatize words
    words = [word for word in words if word.lower() not in stop_words]  # Remove stop words
    return ' '.join(words)


df['prepros_text'] = df['text'].apply(preprocess_text)
# df_sub=df.iloc[:1000,:]
# df_sub['prepros_text'] = df_sub['text'].apply(preprocess_text)
# df_sub
```
